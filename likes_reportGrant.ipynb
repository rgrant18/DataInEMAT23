```python
import pandas as pd
import json
```

```python
with open(r"C:\Users\chris\Downloads\drive-download-20230912T152207Z-001\technogecko_20200714_toshare\likes.json") as cm:
    likes = json.load(cm)
```

```python
liked_posts = pd.DataFrame(likes['media_likes'])
```

```python
liked_posts.iloc[0:100].sort_values(1, ascending = False).groupby(1).count()
```

The dataset above shows a total of 100 likes made by Dr. Silva. These 100 likes are the first 100 indexed from Dr. Silva's account. From this, we are able to infer who's accounts may have the most likes from Dr. Silva in the future. The outliers in this dataset will prove to be the accounts with the most interaction, potential favorite accounts or close friends, and with the least interaction, potential mutual friends.
